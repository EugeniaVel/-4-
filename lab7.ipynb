```python
def euler(T,N,omega,omega0,f):
    dt=T/N
    tt=0
    xx=0
    vv=0
    ans=[[tt,xx]]
    for n in range(N):
        [tt,xx,vv]=[tt+dt,xx+vv*dt,vv+(-omega0^2*xx+f*sin(omega*tt))*dt]
        ans.append([tt,xx])
    return ans
```

```python
n71=euler(50,100,2,1,1)
p71=line(n71, axes_labels=["$t$","$x$"])
p71.save("p71.png")
p71
```
