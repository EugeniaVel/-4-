```python
var("t,x0")
omega=1
x=function("x")(t)
desolve(diff(x,t,2)+omega^2*x==0,x,ics=[0,x0,0])
```

```python
var("x,v")
def euler(F,x0,T,N):
    dt=T/N
    tt=0
    xx=x0
    vv=0
    ans=[[tt,xx]]
    for n in range(N):
        [tt,xx,vv]=[tt+dt,xx+RR(vv*dt),vv+RR(F.subs([t==tt,x==xx,v==vv])*dt)]
        ans.append([tt,xx])
    return ans
fl=-omega^2*x
pe=line(euler(fl,1,8*pi,100), axes_labels=["$t$","$x$"])
pe.save("pe.png")
pe
```

```python
p1=line(euler(fl,1,4*pi,100), axes_labels=["$t$","$x$"])+plot(cos(t),(t,0,4*pi),color="red")
p1.save("p1.png")
p1
```

```python
def euler2(F,x0,T,N):
    dt=T/N
    tt=0
    xx=x0
    vv=0
    ans=[[tt,xx,vv]]
    for n in range(N):
        [tt,xx,vv]=[tt+dt,xx+RR(vv*dt),vv+RR(F.subs([t==tt,x==xx,v==vv])*dt)]
        ans.append([tt,xx,vv])
    return ans
H=v^2/2-omega^2*cos(x)
HH=line([[tt,H.subs([x==xx,v==vv])] for [tt,xx,vv] in euler2(fn,1,pi*4,1000)], axes_labels=["$t$","$H$"])
HH.save("HH.png")
HH
```
