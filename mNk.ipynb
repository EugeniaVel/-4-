```python
def mnk(b,n):
    var('x')
    a=var(['a'+str(i) for i in range(n+1)])
    sk=sum([a[i]*(x^i) for i in range(n+1)])
    f=sum([(sk.subs(x=xx)-y)^2 for [xx,y] in b])
    F=solve([diff(f,i) for i in a],a)
    om=sk.subs(F)
    return SR(RR[x](om))
```
