```python
var("t,x,y")
def euler(N,o):
    dt=1
    t=0
    x=50
    y=20
    a=0.07
    b=0.09
    c=0.002
    d=0.001
    v1=0
    v2=0
    fx=0
    fy=0
    if o==1:
        ans=[[t,x]]
        for n in range(N):
            xi=x
            yi=y
            if x<1:
                [t,x,y]=[t+dt,0,(y+(d*x-b)*y)*dt]
            if y<1:
                [t,x,y]=[t+dt,(x+(a-c*y)*x)*dt,0]
            else:
                [t,fx,x,fy,y]=[t+1,(a-c*y)*x,x+(a-c*y)*x,(d*x-b)*y,y+(d*x-b)*y]
                [x,y]=[xi+(fx+(a-c*y)*x)/2,yi+(fy+(d*x-b)*y)/2]
            ans.append([t,x])
        return ans
    elif o==2:
        ans=[[t,y]]
        for n in range(N):
            xi=x
            yi=y
            if x<1:
                [t,x,y]=[t+dt,0,(y+(d*x-b)*y)*dt]
            if y<1:
                [t,x,y]=[t+dt,(x+(a-c*y)*x)*dt,0]
            else:
                [t,fx,x,fy,y]=[t+1,(a-c*y)*x,x+(a-c*y)*x,(d*x-b)*y,y+(d*x-b)*y]
                [x,y]=[xi+(fx+(a-c*y)*x)/2,yi+(fy+(d*x-b)*y)/2]
            ans.append([t,y])
        return ans
    elif(o==3):
        ans=[[x,y]]
        for n in range(N):
            xi=x
            yi=y
            if x<1:
                [t,x,y]=[t+dt,0,(y+(d*x-b)*y)*dt]
            if y<1:
                [t,x,y]=[t+dt,(x+(a-c*y)*x)*dt,0]
            else:
                [t,fx,x,fy,y]=[t+1,(a-c*y)*x,x+(a-c*y)*x,(d*x-b)*y,y+(d*x-b)*y]
                [x,y]=[xi+(fx+(a-c*y)*x)/2,yi+(fy+(d*x-b)*y)/2]
            ans.append([x,y])
        return ans
    elif(o==4):
        ans=[[x,(a-c*y)*x]]
        for n in range(N):
            xi=x
            yi=y
            if x<1:
                [t,x,y]=[t+dt,0,(y+(d*x-b)*y)*dt]
            if y<1:
                [t,x,y]=[t+dt,(x+(a-c*y)*x)*dt,0]
            else:
                [t,fx,x,fy,y]=[t+1,(a-c*y)*x,x+(a-c*y)*x,(d*x-b)*y,y+(d*x-b)*y]
                [x,y]=[xi+(fx+(a-c*y)*x)/2,yi+(fy+(d*x-b)*y)/2]
            ans.append([x,fx])
        return ans
    elif(o==5):
        ans=[[y,(d*x-b)*y]]
        for n in range(N):
            xi=x
            yi=y
            if x<1:
                [t,x,y]=[t+dt,0,(y+(d*x-b)*y)*dt]
            if y<1:
                [t,x,y]=[t+dt,(x+(a-c*y)*x)*dt,0]
            else:
                [t,fx,x,fy,y]=[t+1,(a-c*y)*x,x+(a-c*y)*x,(d*x-b)*y,y+(d*x-b)*y]
                [x,y]=[xi+(fx+(a-c*y)*x)/2,yi+(fy+(d*x-b)*y)/2]
            ans.append([y,fy])
        return ans
```
