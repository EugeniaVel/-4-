```python
var("v,x,t")
def euler(f,xx,T,N):
    dt=T/N
    vv=0
    tt=0
    rez=[[0,xx]]
    for n in range(N):
        [tt,xx,vv]=[tt+dt,RR(xx+vv*dt),RR(vv+f.subs(t==tt,x==xx,v==vv)*dt)]
        rez.append([n*dt,xx])
    return rez
```

```python
omega=1
f=-omega^2*x
var("t,x0")
fn=-omega^2*sin(x)
p1=line(euler(f,0.2,2*pi,10000),axes_labels=["$t$","$x$"])+line(euler(fn,0.2,2*pi,10000),color="red")
p1.save("p91.png")
p1
```
