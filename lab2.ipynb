```python
def eulersov(f,T,N,x0):
    dt=T/N
    tt=0
    xx=x0
    ans=[[tt,xx]]
    fx=0
    for n in range(N):
        xi=xx
        [tt,xx,fx]=[tt+dt,xx+fx*dt,RR(f.subs(x==xx,t==tt))]
        xx=xi+dt*(fx+RR(f.subs(x==xx,t==tt)))/2
        ans.append([tt,xx])
    return ans
```
