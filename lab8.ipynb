```python
omega0=1
s=9/100
A=matrix([[omega0,s],[s,omega0]])
[lambda1,lambda2]=A.eigenvalues()
omega1=sqrt(lambda1)
omega2=sqrt(lambda2)
L=A.right_eigenvectors()
(xi1,eta1)=L[0][1][0]
(xi2,eta2)=L[1][1][0]
var("t")
x1=xi1*sin(omega1*t)
y1=eta1*sin(omega1*t)
x2=xi2*sin(omega2*t)
y2=eta2*sin(omega2*t)
```

```python
def euler(T,N):
    dt=T/N
    t=0
    x=0
    y=0
    u=xi1*omega1
    v=eta1*omega1
    ans=[[t,x]]
    for n in range(N):
        t=t+dt
        [x,y,u,v]=[x+u*dt,y+v*dt,u-RR(omega0^2*x+s*y)*dt,v-RR(omega0^2*y+s*x)*dt]
        ans.append([t,x])
    return ans
```

```python
p5=line(euler(50,10000))+plot(x1,(t,0,50),axes_labels=["$t$","$x$"],color="red")
p5.save("p85.png")
p5
```
